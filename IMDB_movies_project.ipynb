```python
import mysql.connector
import pandas as pd
connection = mysql.connector.connect(
    host="Hidden for security purposes",
    user="Hidden for security purposes",
    password=r"Hidden for security purposes", 
    database="Hidden for security purposes",
    auth_plugin="mysql_native_password"
)

cursor = connection.cursor()

```

```python
def run_query(query):
    cursor.execute(query)
    rows = cursor.fetchall()
    cols = [i[0] for i in cursor.description]
    return pd.DataFrame(rows, columns=cols)
```

```python
#a) Can you get all data about movies?
run_query("SELECT * FROM movies;")
```

```python
#b) How do you get all data about directors?
run_query("SELECT * FROM directors;")
```

```python
#c) Check how many movies are present in IMDB.
run_query("SELECT COUNT(*) AS total_movies FROM movies;")
```

```python
#d) Find these 3 directors: James Cameron ; Luc Besson ; John Woo
run_query("select name from directors where name IN('James Cameron','Luc Besson','John Woo');")
```

```python
#e) Find all directors with name starting with S.
run_query("select name from directors where name LIKE 's%';")
```

```python
#f) Count female directors.
run_query("select count(*) from directors where gender='1';")
```

```python
#g) Find the name of the 10th first women directors?
run_query("SELECT name FROM directors WHERE gender = 1 ORDER BY name LIMIT 1 OFFSET 9;")
```

```python
#h) What are the 3 most popular movies?
run_query("SELECT title FROM movies ORDER BY popularity DESC LIMIT 3;")
```

```python
#i) What are the 3 most bankable movies?
run_query("SELECT title,revenue FROM movies ORDER BY revenue DESC LIMIT 3;")
```

```python
#j) What is the most awarded average vote since the January 1st, 2000?
run_query("SELECT title,vote_average FROM movies WHERE release_date >= '2000-01-01' ORDER BY vote_average DESC LIMIT 1;")
```

```python
#k) Which movie(s) were directed by Brenda Chapman?
run_query("SELECT m.title FROM movies m JOIN directors d ON m.uid = d.uid WHERE d.name = 'Brenda Chapman';")
#Brenda Chapman’s ID and UID exists in the directors table, but it does not appear in the movies table, so no movies are assigned to her.
```

```python
#l) Which director made the most movies?
run_query("SELECT d.name AS director_name,COUNT(m.id) AS total_movies,GROUP_CONCAT(m.title SEPARATOR ', ') AS movie_list FROM directors d JOIN movies m ON d.id = m.director_id GROUP BY d.id, d.name ORDER BY total_movies DESC LIMIT 1;")
```

```python
#m) Which director is the most bankable?
run_query("SELECT d.name, SUM(m.revenue) AS total_revenue FROM directors d JOIN movies m ON d.id = m.director_id GROUP BY d.name ORDER BY total_revenue DESC LIMIT 1;")
```
